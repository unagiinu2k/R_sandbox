## チルダを付けてモデルを列に追加

チルダの意味は
https://stackoverflow.com/questions/44834446/what-is-meaning-of-first-tilde-in-purrrmap

```r
by_country = by_country  %>%  mutate(model = purrr::map(data  ,~ lm(lifeExp ~ year , data = .)))
```
